### Machine Leaning - SVM
---

The goal of SVM is:
- Find the optimal separating **hyperplane** that **maximizes the margin** of the train data.
- SVM fits a decision boundary similarly to a regression, but uses a different loss function that is called the "hinge loss" (as opposed to the log loss in logistic regression).
- SVMs are notorious for being less intuitive than other classifiers such as kNN and Logistic regression (lots of linear algebra, geometry and more)

### Machine Leaning - SVM
---
- Given a particular hyperplane (as the black line from the previous graph) we can calculate the distance between the hyperplane and the closest data point. Then we can double this distance and the final value will be called **Margin**
- In this**Margin** area, we must make sure that there are not any data points (but this will be an issues when we have noisy data).
- From the following examples we can conclude:
    - If a hyperplane is very close to a data point, then we will have a **small margin**
    - The further a hyperplane is from a data point, then we will have a **larger margin** (less able to MISCLASSIFY AN OBSERVATION)
    
- **The optimal hyperplane will be the one with the larger margin** which is the objective of SVM:
    - **Find the optimal separating hyperplane that maximizes the margin of the train data.**
    - This Large Margin, plays a significant role when the classification of observations is "linearly separable"
    - In here SVM fits the **decision boundary,** that is defined by the largest margin between the closest points for each class 
    - This is commonly called the **"maximum margin hyperplane (MMH)"**

### Machine Leaning - SVM
---
#### Finding the optimal hyperplane
- Notice that we are assuming that the classification of observations is **linearly separable.** It is in here where SVM fits the **the decision boundary** which is the largest margin between the closest points for each class => this is AKA as the **maximum margin hyperplane (MMH)**


- The points SVM uses to fit the decision boundary are called **"support vectors".** This term comes from linear algebra: in a vector space, points can be defined as a vector between the origin and that point. (we have already discuss this above)


- The question we should be asking now is: Why we need to maximize this Margin? 
    -This is because **SVM solves for the decision boundary that minimizes generalization error.**


- Observations that are near the decision boundary between the classes are the most ambiguous observations. They are the observations that are approaching equal chance to be one class or the other.


- SVM, instead of considering all the observations "equally" in the loss function it **minimizes**, defines it's fit using the most ambiguous points. It's decision boundary is _safe_ in that errors in new measured observations are not likely to cause the SVM to mis-classify.


- The SVM is concerned with generalization to new data.

### Machine Leaning - SVM
---
#### Maximum margin hyperplane function
- The decision boundary (which is equivalent to the MMH) is derived by the [discriminant function](https://en.wikipedia.org/wiki/Discriminant_function_analysis#Discriminant_functions). A linear combination of predictors that maximizes the difference between groups. 

    - At this poit you should be already familiar with the discriminant function for logistic regression, represented by the sigmmoid curve solved by the log loss, from previous econometric lectures (not covering this point in this class) 
    

- In an SVM, the discriminant function is:

### $$ f(x) = sign(w^T x + b) $$

... where **$w$** is the normalized weight vector, perpendicular to the decision boundary.

... $b$ is the "bias", which is the corollary (a natural consequence) to the intercept in regression.

- (Note that the $bias$ term is sometimes called the intercept. This has nothing to do with bias-variance tradeoff)



- The solution to $f(x)$ for an $x$ observation is a sign (-1, or 1), and this determines the (binary-case) class label of the observation $x$.

### Machine Leaning - SVM
---
#### The hinge - loss function
- It’s convex like the exponential and logistic loss: optimization is fairly easy
- It grows slower below zero: the penalty for misclassification is more mild
- It is zero after 1: once correct and confident enough, there is no bonus for driving the function higher
- It’s non-differentiable at the hinge point (we can not classify accurately anymore when we get to the hinge point: simple gradient doesn’t work

### Machine Leaning - SVM
---
#### The hinge - loss function

- Is the function SVM minimizes to find the boundary:

### $$  minimize \;\; ||w||^2 + C\sum_{i=1}^N max\left(0, 1 - y_i \: f(x_i)\right) $$

... where $||w||^2$ is the magnitude of the weight vector squared

... $C$ is a regularization term where a small $C$ creates a **wider margin**, and large $C$ creates a **tighter margin**.

### $$\sum_{i=1}^N max\left(0, 1 - y_i \: f(x_i)\right)$$ 

... is the sum of the errors. 
- Remember from the nature of the curve:
    - If $f(x_i) > 1$, the point lies _outside_ the margin and does not contribute to the loss.

    - If $f(x_i) = 1$ the point is _on_ the margin and does not contribute to the loss.

    - If $f(x_i) < 1$ the point lies _inside_ the margin and **does** contribute to the loss.

### $$  minimize \;\; ||w||^2 + C\sum_{i=1}^N max\left(0, 1 - y_i \: f(x_i)\right) $$


The right hand side of the function to be optimized is saying:

> Every point should be on the correct side of the margins for both classes.
    
The left hand side is saying:

> Maximize the distance between the margins by modifying the weight vector.

It seems in the above equation like we are minimizing the margin due to the $||w||^2$ term. This is in fact maximizing the margin, but the math behind it is more involved. [This page, does a good job explaining this.](http://nlp.stanford.edu/IR-book/html/htmledition/support-vector-machines-the-linearly-separable-case-1.html)

### Machine Leaning - SVM
---
#### The hyperparameter C
- The hyper-parameter $C$ controls the extent to which the SVM is tolerant to errors. It is sometimes called the **"soft-margin constant".** 


- $C$ affects the strength of the "penalty" similar to the lambda terms in the Ridge and Lasso. 


- By multiplying the sum of the errors, which are the distances from the margins to the points inside of the margin, it allows the SVM to classify non-linearly separable problems by **allowing errors to occur.** 


- The lower the value of $C$, the more misclassified observations errors are allowed. These misclassified points are known as **"slack variables".** Reducing the effect of errors on the loss function puts the emphasis on widening the margin.


[See this great tutorial for more math and from where I got some the ideas for this lecture.](http://www.svm-tutorial.com/2015/06/svm-understanding-math-part-3/#more-457)

### Machine Leaning - SVM

#### The kernel trick to solve non-linearly separable problems 

- The math behind the "kernel trick" is beyond the scope of this lecture, but it allows an SVM to classify non-linearly separable problems. **It is a big reason why SVMs are so popular.**


- What we need to know is that with an SVM **one can arbitrarily transform observations that _have no linear seperability_ by putting them into a different "dimensional space".** 


- This is done by "wrapping" the predictors in a kernel function that transforms them to a different dimensional space. At the end, they can be transformed back, along with the decision boundary, to their original dimensional space.

[Check out these lecture slides for a good overview.](http://www.robots.ox.ac.uk/~az/lectures/ml/lect3.pdf)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('fivethirtyeight')

from ipywidgets import *
from IPython.display import display

from sklearn import svm, datasets

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

'''Using the Iris dataset for this example'''
iris = datasets.load_iris()
X = iris.data[:, :2] # we only take the first two features. We could
y = iris.target
```

```python
## We do not scale our data since we want to plot the support vectors
## www.analyticsvidhya.com

''' Using the linear kernel'''
C = 1.0 # SVM regularization parameter
svc = svm.SVC(kernel='linear', C=1, gamma='auto').fit(X, y)
fig = plt.figure(figsize=(18,6))

'''create a place holder for the plot'''
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
 np.arange(y_min, y_max, h))

'''defining subplots'''
plt.subplot(1, 2, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)

'''printing the graph'''
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with linear kernel')

''''''''''''''''''''''''''''''
''' Using the rbf kernel'''
C = 1.0 # SVM regularization parameter
svc = svm.SVC(kernel='rbf', C=1, gamma='auto').fit(X, y)

'''create a place holder for the plot'''
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
 np.arange(y_min, y_max, h))

'''defining subplots'''
plt.subplot(1, 2, 2)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)

'''printing the graph'''
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with rbf kernel')



plt.show()
```
